```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
names = ['Timestamp','Gender','Age','Last sleep','Sufficient','More sleep','Less sleep days','Need model']

df = pd.read_csv('Sleep Survey.csv',names = names)
df = df.drop('Timestamp',axis=1)
df = df.drop(0,axis = 0)
df
```

```python
# GENDER STATS
male = len(df[df["Gender"]=='Male'])
female = len(df[df["Gender"]=='Female'])
print(male)
print(female)
```

```python
slept_less = df[df["Sufficient"]=='No']
slept_less

more_sleep = list(slept_less["More sleep"])
slept_less_more_yes = more_sleep.count("Yes")
slept_less_more_no = more_sleep.count("No")
slept_less_more_maybe = more_sleep.count("Maybe")

cent = slept_less_more_yes*100/len(slept_less)
print("People who slept less and think they need more sleep =",cent,"%")

plt.pie([slept_less_more_yes,slept_less_more_no,slept_less_more_maybe],colors=['gold','mediumpurple','mediumaquamarine'],labels=['yes','no','maybe'], autopct='%1.1f%%', explode=[0.05,0.05,0.05])
plt.title("Slept less and need more sleep in next days")
plt.plot()
```

```python
# Total number of people who think they need more sleep

male_data = df[df["Gender"]=='Male']
female_data = df[df["Gender"]=='Female']

male_need_more_sleep = len(male_data["More sleep"]=='Yes')
print(male_need_more_sleep)

female_need_more_sleep = len(female_data["More sleep"]=='Yes')
print(female_need_more_sleep)
```

```python
# Number of days when people get less sleep than required

days12 = len(df[df["Less sleep days"]=='1-2'])
days36 = len(df[df["Less sleep days"]=='3-6'])
days7 = len(df[df["Less sleep days"]=='Everyday'])

plt.pie([days12,days36,days7],colors=['gold','mediumpurple','mediumaquamarine'],labels=['1-2 days','3-6 days','Everyday'],autopct='%1.1f%%', explode=[0.05,0.05,0.05])
plt.title("People who think they don't get required sleep")
plt.show()
```

```python
# MODEL REQUIREMENT STATS

need = list(df['Need model'])
yes,no,maybe = need.count('Yes'),need.count('No'),need.count("Maybe, I don't know.")
plt.pie([yes,no,maybe],colors=['gold','mediumpurple','mediumaquamarine'],labels=['yes','no','maybe'], autopct='%1.1f%%', explode=[0.05,0.05,0.05])
plt.title("Need the model")
plt.plot()
```
